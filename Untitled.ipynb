```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\project\6 2\Sales_data 1.csv")
df
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Total Sales"], bins=20, kde=True, color="blue")

# Customize the plot
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.grid(True)
```

```python
plt.figure(figsize=(10, 7))
sns.countplot(x=df["Product Category"], palette="viridis")

# Customize the plot
plt.title("Sales Count by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate labels for better visibility
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Quantity Sold"], y=df["Total Sales"], color="yellow", alpha=0.7)

# Customize the plot
plt.title("Scatter Plot: Quantity Sold vs Total Sales")
plt.xlabel("Quantity Sold")
plt.ylabel("Total Sales")
plt.grid(True)

# Show the plot
plt.show()
```
